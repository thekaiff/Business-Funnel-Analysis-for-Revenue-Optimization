# **Funnel Overview:**

**LEVEL 1: Awareness**
- this stage perticularly shows the potential buyer becomes aware of the product or services. 
- expl: a clothing brand creating awareness thro various medium by creating ads

**LEVEL 2: Consideration**
- in this stage the potential buyer becomes interested in the products or services you are rendering.
this is after they have heard about your brand, now they want to know more about the product/services
- expl: buyer wants to know what type of cloths this brand is selling, do they have his/her size or 
his/her type of choice 
- in this stage the buyer also potentially evaluates your products and services with respect to their
needs, whether it suits them or not

**LEVEL 3: Conversion** (most critical stage)
- this stage simply push the buyer to take actions and purchase the products. Here's where you make money 
from as a business

**LEVEL 4: Loyalty**
- in this stage, the buyers becomes our loyal customer. they love our brand, our services, our products.
And are willing to buy more from us

**LEVEL 5: Advocacy**
- in this stage they become like our advocate, they tell they friends/family about our products.
basically helps in marketing our products

This means, if buyers reach level 4 and level 5, they will be the one that will iteragting this process for us again!
